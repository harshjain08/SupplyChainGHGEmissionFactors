# Data Analytics Project
This notebook covers the initial data exploration and analysis of the dataset provided.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload the file
uploaded = files.upload()

# Load the dataset
file_path = 'SupplyChainGHGEmissionFactors_v1.2_NAICS_CO2e_USD2021.csv'  # The file will be uploaded to the current directory
df = pd.read_csv(file_path)
```

## 1. Basic Data Exploration

```python
def data_exploration(df):
    print("First 5 rows of the dataset:")
    display(df.head())

    print("\nDataset Information:")
    df.info()

    print("\nSummary Statistics:")
    display(df.describe())

    print("\nChecking for missing values:")
    display(df.isnull().sum())

data_exploration(df)
```

## 2. Data Cleaning

```python
def data_cleaning(df):
    # Handle missing values
    df = df.dropna()  # Dropping rows with missing values, adjust as necessary

    # Convert columns to appropriate data types if necessary
    # Example: df['column_name'] = df['column_name'].astype('int')

    print("Data after cleaning:")
    display(df.head())

    return df

df = data_cleaning(df)
```

## 3. Data Visualization

```python
def data_visualization(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x='Supply Chain Emission Factors with Margins', bins=20, kde=True)
    plt.title('Distribution of Supply Chain Emission Factors with Margins')
    plt.show()


data_visualization(df)
```

```python
def top_10_bar_plot(df):
    top_10_emission_factors = df.nlargest(10, 'Supply Chain Emission Factors with Margins')[['2017 NAICS Title', 'Supply Chain Emission Factors with Margins']]
    plt.figure(figsize=(12, 6))
    sns.barplot(data=top_10_emission_factors, x='Supply Chain Emission Factors with Margins', y='2017 NAICS Title', palette='viridis')
    plt.title('Top 10 Industries by Emission Factors')
    plt.xlabel('Emission Factors with Margins')
    plt.ylabel('Industry')
    plt.show()

top_10_bar_plot(df)
```

```python
def box_plot_emission_factors(df):
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df, x='Supply Chain Emission Factors with Margins')
    plt.title('Box Plot of Supply Chain Emission Factors with Margins')
    plt.xlabel('Emission Factors with Margins')
    plt.show()

box_plot_emission_factors(df)
```

```python
def pair_plot(df):
    sns.pairplot(df[['Supply Chain Emission Factors without Margins', 'Margins of Supply Chain Emission Factors', 'Supply Chain Emission Factors with Margins']])
    plt.show()

pair_plot(df)
```

```python
def histogram_margins(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x='Margins of Supply Chain Emission Factors', bins=20, kde=True)
    plt.title('Distribution of Margins of Supply Chain Emission Factors')
    plt.xlabel('Margins of Supply Chain Emission Factors')
    plt.show()

histogram_margins(df)
```

## 4. Basic Data Analysis

```python
def data_analysis(df):
    top_10_emission_factors = df.nlargest(10, 'Supply Chain Emission Factors with Margins')[['2017 NAICS Title', 'Supply Chain Emission Factors with Margins']]
    print("Top 10 industries with highest emission factors:")
    display(top_10_emission_factors)

    emission_by_naics = df.groupby('2017 NAICS Title')['Supply Chain Emission Factors with Margins'].mean().sort_values(ascending=False).head(10)
    print("Average emission factors by NAICS Title:")
    display(emission_by_naics)

data_analysis(df)
```
